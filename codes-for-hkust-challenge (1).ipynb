```python
import numpy as np
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filepath='../input/temperature-daliy/temperature_daily.csv'
data1=pd.read_csv(filepath)
df=pd.DataFrame(data1)
df['date']=pd.to_datetime(df['date'])
df['month']=df['date'].dt.month
df['year']=df['date'].dt.year
mdt=df.groupby(['year','month']).agg(
    max_temp=('max_temperature', 'max'),
    min_temp=('min_temperature', 'min')
).reset_index()
mdtmax=mdt.pivot_table(index='month',columns='year',values='max_temp')
mdtmin=mdt.pivot_table(index='month',columns='year',values='min_temp')
```

```python
plt.figure(figsize=(20,6))
plt.title("Average Month Temperature ,by Year")
ax=sns.heatmap(data=mdtmax, annot=True,cmap='hot_r',vmin=4,vmax=38)
plt.xlabel('year')
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

plt.figure(figsize=(20,6))
plt.title("Average Month Temperature ,by Year")
bx=sns.heatmap(data=mdtmin, annot=True,cmap='hot_r',vmin=4,vmax=38)
plt.xlabel('year')
bx.xaxis.tick_top()
bx.xaxis.set_label_position('top')
```

```python
import pandas as pd
import plotly.graph_objects as go

# 读取数据
df = pd.read_csv(filepath)
df['date'] = pd.to_datetime(df['date'])
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month

# 计算月平均最高和最低温度
monthly_high = df.groupby(['year','month'])['max_temperature'].mean().reset_index()
monthly_low = df.groupby(['year','month'])['min_temperature'].mean().reset_index()

# 创建热图矩阵
pivot_high = monthly_high.pivot(index='month', columns='year', values='max_temperature')
pivot_low = monthly_low.pivot(index='month', columns='year', values='min_temperature')

# 创建交互式图表
fig = go.Figure()

# 添加最高温度热图
fig.add_trace(go.Heatmap(
    z=pivot_high,
    x=pivot_high.columns,
    y=pivot_high.index,
    colorscale='hot_r',
    name='最高温度',
    hovertemplate='年份: %{x}<br>月份: %{y}<br>温度: %{z:.1f}°C'
))

# 添加最低温度热图
fig.add_trace(go.Heatmap(
    z=pivot_low,
    x=pivot_low.columns,
    y=pivot_low.index,
    colorscale='Viridis',
    name='最低温度',
    visible=False,
    hovertemplate='年份: %{x}<br>月份: %{y}<br>温度: %{z:.1f}°C'
))

# 添加切换按钮
fig.update_layout(
    title='香港月平均温度（2000-2025）',
    updatemenus=[
        dict(
            buttons=[
                dict(label='最高温度', method='update',
                     args=[{'visible': [True, False]}, {'title': '月最高温度'}]),
                dict(label='最低温度', method='update',
                     args=[{'visible': [False, True]}, {'title': '月最低温度'}])
            ],
            direction='down',
            pad={'r':10,'t':10},
            showactive=True,
            x=0.1,
            xanchor='left',
            y=1.1,
            yanchor='top'
        )
    ],
    coloraxis_colorbar=dict(title='温度 (°C)')
)

fig.show()
```
